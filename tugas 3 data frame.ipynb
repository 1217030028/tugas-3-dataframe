```python
import pandas as pd
import numpy as np

# Membuat array x menggunakan NumPy
x = np.arange(0, 30) 

# Menghitung nilai sinus dan cosinus dari setiap elemen dalam array x
y = np.sin(x)
z = np.cos(x)

# Membuat DataFrame dengan pandas
dataframe = pd.DataFrame({
    'x': x, 
    'y': y
})

# Menambahkan kolom 'z' ke DataFrame
dataframe['z'] = z

# Menampilkan DataFrame
dataframe
```
